# How many songs are released every year ?

```python
#Extracting year of release of a song from date and making a separate data frame containing year of release and number of songs released.
tracks['year']= pd.DatetimeIndex(tracks.index).year
Per_year=tracks.groupby('year')[['name']].count()
Per_year.reset_index(inplace=True)

#Plotting a bar graph showing varaition of number of songs vs year
sb.set_style(style='darkgrid')
plt.figure(figsize=(20,7))
plt.xticks(rotation=90)

sb.barplot(x='year', y='name', data= Per_year).set(title='No. of Songs vs Year')
plt.ylabel('Number of Songs', fontsize=16)
plt.xlabel('Year', fontsize=16)
```

### We can see clearly that from 1900 to 2021 number of songs released every year is increasing consistently.

# How are these parameters of each song related to each other ?

# Correlation HeatMap

1. This Heatmap, is a graphical representation of of a correlation matrix representing the correlation between different variables.
2. Here, as we can see different parameters of a song are displayed in the correlation matrix, and their correlation with other parameters is displayed inside the box, with a number and colour of the box.
3. Lighter color (or positive correlation coefficient) means positive correlation exists between two variables, A positive correlation is a relationship between two variables that tend to move in the same direction. A positive correlation exists when one variable tends to decrease as the other variable decreases, or one variable tends to increase when the other increases
4. Darker Color (or negative correlation coefficient) means that a negative correlation exists between two variables such that as the value of one variable increases, the other decreases.


```python
correlation_tracks=tracks.drop(['mode','key','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(20,10))
heatmap=sb.heatmap(correlation_tracks, annot=True , fmt='.1g',vmin=-1 ,cmap='magma', linewidth=1, linecolor='black')
heatmap.set_title('Correlation HeatMap')
```

# Observations

1. In the above heatmap, we can figure out the correlations of various parameters with each other.
2. For example, energy and loudness have a positive correlation, for high loudness, high energy was observed(correlation coefficient +0.8).
3. Acousticness and energy have a negative correlation, for high acousticness low energy was observed (correlation coefficient -0.7).
4. Popularity and Loudness have a moderately positive correlation, (correlation coefficient +0.3)

# Regression Plot

1. To get an idea of Regression Plot it's better to understand Regression Analysis. Regression Analysis, a statistical technique, is used to evaluate the relationship between two or more variables. Regression analysis helps an organisation to understand what their data points represent and use them accordingly with the help of business analytical techniques in order to do better decision-making.
2. Regression plots as the name suggests creates a regression line between 2 parameters and helps to visualize their linear relationships. 


# How is loudiness of song related to the energy of the song ?

```python
# This regression plot has been made using the sample dataset generated above using only 0.4% of the data.
plt.figure(figsize=(15,7))
sb.regplot(data=sample, y= 'loudness', x='energy', color= 'r').set(title='Loudness vs Energy Correlation')
```

# Observations 

1. The above plot is a graphical representation of the correlation between Loudness of a song and the energy.
2. As seen in the above heatmap, the coefficent of correlation for this pair was +0.8, i.e., strong positive correlation exists between them.
3. It can be seen clearly in this regression plot that when loudness of a song is low it's energy is also low, and vice versa.

# How is loudiness of song related to the popularity of the song ?

```python
# This regression plot has been made using the sample dataset generated above using only 0.4% of the data.
plt.figure(figsize=(15,7))
sb.regplot(data=sample, y= 'energy', x='acousticness', color= 'c').set(title=' Energy vs Acousticness Correlation')
```

# Observations 

1. The above plot is a graphical representation of the correlation between acousticness of a song and the energy.
2. As seen in the above heatmap, the coefficent of correlation for this pair was -0.7, i.e., strong negative correlation exists between them.
3. It can be seen clearly in this plot that when acousticness of a song is low it corresponds to high energy, and vice versa.

#### We can see that there are two Children's Music Genres that are being displayed above, this should be fixed, as currently the value of total unique genres is inaccurate.

```python
genre_df.genre.replace("Children’s Music", "Children's Music", inplace=True)
Unique=genre_df['genre'].unique()
print('Number of different unique genres are: ', len(Unique),'\n',Unique)
```

# Popularity of different genres.

```python
#Finding popularity of different genres, by creating a dataframe of popularity and genres.
Popular=genre_df.groupby('genre')[['popularity']].sum().sort_values('popularity', ascending = False)
Popular.reset_index(inplace=True)

#Plotting a bargraph of popularity vs Genres
sb.set_style(style='darkgrid')
plt.figure(figsize=(20,7))
plt.xticks(rotation=90)
sb.barplot(x='genre', y='popularity', data= Popular).set(title='Popularity of different Genres')
```

# What are the top 5 most popular genres ?

```python
#Creating a new dataframe with just genre and it's total popularity to find out top 5 most popular genre on spotify.
top_5=genre_df.groupby('genre')[['popularity']].sum()
top_5=top_5.sort_values('popularity', ascending=False).head(5)

#Resetting index of top_5 dataframe 
top_5.reset_index(inplace=True)

#plotting bargraph for popularity vs genre
sb.set_style(style='darkgrid')
plt.figure(figsize=(15,7))
sb.barplot(x='genre', y='popularity', data= top_5)
```

```python
a=top_5['genre'].head(5)
print('Top 5 genres on spotify are:\n',a)
```
